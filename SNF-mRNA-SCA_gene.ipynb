```python

import pandas as pd
import numpy as np
from snf import compute
from sklearn import cluster
from sklearn.metrics import silhouette_score
import umap.umap_ as umap
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Step 1: Load your CSVs
# -------------------------------
data1 = pd.read_csv("SCNA_gene(in).csv", index_col=0).transpose()  # rows = patients
data2 = pd.read_csv("mRNA(in).csv", index_col=0).transpose()

# Align patient IDs
data2 = data2.loc[data1.index]

data_list = [data1.values, data2.values]

# -------------------------------
# Step 2: Parameter tuning for K and mu
# -------------------------------
K_values = [10, 15, 20, 25]   # number of neighbors
mu_values = [0.3, 0.5, 0.7]   # scaling parameter

best_silhouette = -1
best_params = None
best_fused = None
best_labels = None

for K in K_values:
    for mu in mu_values:
        # Affinity matrices
        affinities = compute.make_affinity(data_list, K=K, mu=mu)
        
        # Fuse networks
        fused = compute.snf(affinities, K=K)
        
        # Cluster samples: using 2 clusters (adjust if you want)
        labels = cluster.spectral_clustering(fused, n_clusters=2)
        
        # Silhouette score: distance = 1 - similarity
        distance_matrix = 1 - fused
        np.fill_diagonal(distance_matrix, 0)
        sil = silhouette_score(distance_matrix, labels, metric="precomputed")
        
        print(f"K={K}, mu={mu} -> silhouette={sil:.4f}")
        
        if sil > best_silhouette:
            best_silhouette = sil
            best_params = (K, mu)
            best_fused = fused
            best_labels = labels

print(f"\nBest parameters: K={best_params[0]}, mu={best_params[1]} -> silhouette={best_silhouette:.4f}")

# -------------------------------
# Step 3: UMAP visualization
# -------------------------------
# Convert fused similarity to distance
distance_matrix = 1 - best_fused
np.fill_diagonal(distance_matrix, 0)

# UMAP embedding
reducer = umap.UMAP(metric="precomputed", n_neighbors=15, min_dist=0.1, random_state=42)
embedding = reducer.fit_transform(distance_matrix)

embedding_df = pd.DataFrame({
    "UMAP1": embedding[:,0],
    "UMAP2": embedding[:,1],
    "Cluster": best_labels,
    "Patient": data1.index
})

# Plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=embedding_df, x="UMAP1", y="UMAP2", hue="Cluster", palette="tab10", s=50)
plt.title(f"UMAP of SNF clusters (K={best_params[0]}, mu={best_params[1]}, silhouette={best_silhouette:.4f})")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.legend(title="Cluster")
plt.tight_layout()
plt.savefig("example_plot.png4", dpi=300, bbox_inches='tight')  # PNG file

plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.inspection import permutation_importance

# ------------------------------
# Step 1: Encode categorical features
# ------------------------------
categorical_cols = ["sex", "participant_country", "tumor_site",
                    "tumor_necrosis", "lymph_vascular_invasion",
                    "perineural_invasion", "additiol_pathologic_findings",
                    "alcohol_consumption", "tobacco_smoking_history"]

for col in categorical_cols:
    if col in df_thresh.columns:
        df_thresh[col] = LabelEncoder().fit_transform(df_thresh[col].astype(str))

# ------------------------------
# Step 2: Prepare features and target
# ------------------------------
feature_cols = ["Cluster", "age", "tumor_size_cm", "bmi"] + categorical_cols
feature_cols = [c for c in feature_cols if c in df_thresh.columns]

X = df_thresh[feature_cols]
y = df_thresh["vital_status"].astype(int)

# ------------------------------
# Step 3: Train/test split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ------------------------------
# Step 4: Scale features (important for KNN)
# ------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ------------------------------
# Step 5: KNN classifier
# ------------------------------
knn = KNeighborsClassifier(n_neighbors=5)  # you can tune n_neighbors
knn.fit(X_train_scaled, y_train)

# ------------------------------
# Step 6: Predict & evaluate
# ------------------------------
y_pred = knn.predict(X_test_scaled)
y_proba = knn.predict_proba(X_test_scaled)[:,1]

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ------------------------------
# Step 7: Feature importance via permutation
# ------------------------------
perm_importance = permutation_importance(knn, X_test_scaled, y_test, n_repeats=30, random_state=42)
importance_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": perm_importance.importances_mean
}).sort_values(by="Importance", ascending=False)

print("Feature importance (permutation):")
print(importance_df.head(10))
```
